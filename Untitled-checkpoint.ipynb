# MVP预测
## 模型一：岭回归

确定和筛选用于MVP预测的列

```python
predictors = ["Age", "G", "GS", "MP", "FG", "FGA", 'FG%', '3P', '3PA', '3P%', '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', 'W', 'L', 'W/L%', 'GB', 'PS/G', 'PA/G', 'SRS']
```

```python
#定义测试(2021年数据)和训练(2021年数据除外)

train = stats[~(stats["Year"] == 2021)]
test = stats[stats["Year"] == 2021]
```

```python
#Ridge_Regression：岭回归
reg = Ridge(alpha=0.1)
reg.fit(train[predictors], train["Share"])#训练predictors的各个数值和share的关系
```

```python
predictions = reg.predict(test[predictors])#拿2021年的数据来测试训练结果
predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
```

```python
combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)#将Player、Share、predicitons（预测结果）组合在一起
combination.sort_values("Share", ascending=False).head(10)#以share的大小进行排序
```

可以看出，通过预测得到的predictions值最大的是字母哥（Giannis Antetokounmpo），也就是我们通过训练后来测试得到的2021年MVP获得者应该会是字母哥（Giannis Antetokounmpo），但实际上最终获得MVP称号的是约老师（Nikola Joki）,他的share值最高，为0.961。可见最终模型预测得到的结果并不符合实际情况，原因是在实际的评选过程中需要考虑的主观因素还有很多，如：球员的招黑程度/舆论风向，是否抱团组成三巨头，是否加入了一个已经具有夺冠实力的球队等；球员打球是否劲爆/飘逸，具有观赏性，狂造犯规骗罚球假摔大家都不喜欢，努力偏执场场绝杀肯定占优势；球员赛场外的表现，领导力、慈善等等。所以我们通过预测只能基于超级客观的数据来预测，很多时候会有很多球迷的主观因素干扰到MVP的结果。

## 识别错误的度量

```python
mean_squared_error(combination["Share"], combination["predictions"]) #求均方误差MSE（mean squared error）来衡量我们的预测值和真实值的差异。
```

可以看出求得的均方误差很小,说明预测模型描述实验数据具有很好的精确度。

```python
combination["Share"].value_counts() #计算combination["Share"]有多少个不同值，并计算每个不同值有在该列中有多少重复值。
```

```python
combination = combination.sort_values("Share", ascending=False)
combination["Rk"]=list(range(1,combination.shape[0]+1))
combination.head(10) 
```

```python
combination=combination.sort_values("predictions", ascending=False)
combination["Predicted_Rk"] = list(range(1, combination.shape[0]+1))
combination.head(10)
```

```python
combination.sort_values("Share", ascending=False).head(10)#把实际RK和预测RK放在一起对比
```

可以看出，预测得到的Predicted_Rk最高的是Giannis Antetokounmpo（字母哥），而字母哥当年实际投票率是排第4。当年实际上的MVP获得者约老师的预测得到的Predicted_Rk为3，可见模型预测的还是稍准的，之所以字母哥当年没拿到MVP，可能是因为他垫脚的事儿吧，这事儿太突然了，模型是当然预测不到的，球迷就不买账了（虽然他2021年的个人数据是真的好）

```python
def find_ap(combination):
    actual = combination.sort_values("Share", ascending=False).head(5)
    predicted = combination.sort_values("predictions", ascending=False)
    ps = []
    found = 0
    seen = 1
    for index, row in predicted.iterrows():
        if row["Player"] in actual["Player"].values:
            found += 1
            ps.append(found / seen)
        seen += 1
        
    return sum(ps) / len(ps)
```

```python
ap = find_ap(combination)
ap
```

## 采用回溯测试预测每年

```python
years = list(range(1991,2022))
aps = []
all_predictions = []

for year in years[5:]: #year[5]是1996年
    train = stats[stats["Year"]<year]
    test = stats[stats["Year"]==year]
    reg.fit(train[predictors], train["Share"])
    predictions = reg.predict(test[predictors])
    predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
    combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
    all_predictions.append(combination)
    aps.append(find_ap(combination))
```

```python
sum(aps) / len(aps)
```

```python
def add_ranks(combination):
    combination = combination.sort_values("Share", ascending=False)
    combination["Rk"] = list(range(1,combination.shape[0]+1))
    combination = combination.sort_values("predictions", ascending=False)
    combination["Predicted_Rk"] = list(range(1,combination.shape[0]+1))
    combination["Diff"] = (combination["Rk"] - combination["Predicted_Rk"])
    return combination
```

```python
ranking = add_ranks(all_predictions[1])
ranking[ranking["Rk"]<5].sort_values("Diff",ascending=False)
```

```python
def backtest(stats, model, year, predictors):
    aps = []
    all_predictions = []
    for year in years[5:]:
        train = stats[stats["Year"] < year]
        test = stats[stats["Year"] == year]
        model.fit(train[predictors],train["Share"])
        predictions = reg.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
        combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
        combination = add_ranks(combination)
        all_predictions.append(combination)
        aps.append(find_ap(combination))
    return sum(aps) / len(aps), aps, pd.concat(all_predictions)
```

```python
mean_ap, aps, all_predictions = backtest(stats, reg, years[5:], predictors)
mean_ap
```

## 模型二：随机森林模型

```python
#建立随机森林模型，并进行测试

rf = RandomForestRegressor(n_estimators=50, random_state=1, min_samples_split=5)

mean_ap, aps, all_predictions1 = backtest(stats, rf, years[28:], predictors + ["NPos", "NTm"])
mean_ap
```

```python
mean_ap, aps, all_predictions1 = backtest(stats, reg, years[28:], predictors)
mean_ap
```

```python
sc = StandardScaler()#标准化

def backtest(stats, model, years, predictors):
    aps = []
    all_predictions = []
    for year in years:
        train = stats[stats["Year"] < year].copy()
        test = stats[stats["Year"] == year].copy()
        sc.fit(train[predictors])
        train[predictors] = sc.transform(train[predictors])
        test[predictors] = sc.transform(test[predictors])
        model.fit(train[predictors],train["Share"])
        predictions = model.predict(test[predictors])
        predictions = pd.DataFrame(predictions, columns=["predictions"], index=test.index)
        combination = pd.concat([test[["Player", "Share"]], predictions], axis=1)
        combination = add_ranks(combination)
        all_predictions1.append(combination)
        aps.append(find_ap(combination))
    return sum(aps) / len(aps), aps, pd.concat(all_predictions1)
```

```python
mean_ap, aps, all_predictions1 = backtest(stats, reg, years[28:], predictors)
all_predictions1[all_predictions1["Rk"] < 3] #随机森林预测结果与实际对比
```

可以看到随机森林的预测结果和岭回归的结果完全相同！
